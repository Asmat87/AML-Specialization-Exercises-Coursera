```python
from preprocessed_mnist import load_dataset
X_train, y_train, X_val, y_val, X_test, y_test = load_dataset()
print(X_train.shape, y_train.shape)
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(X_train[0], cmap="Greys");
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

X_train = X_train.reshape(X_train.shape[0], -1).astype('float32')
X_val = X_val.reshape(X_val.shape[0], -1).astype('float32')
X_test = X_test.reshape(X_test.shape[0], -1).astype('float32')

m, n = X_train.shape
n_classes = len(set(y_train))

sess = tf.Session()
y_train = sess.run(tf.one_hot(y_train, n_classes))
y_test = sess.run(tf.one_hot(y_test, n_classes))
y_val = sess.run(tf.one_hot(y_val, n_classes))
sess.close()

# X_train = X_train/255 - 0.5
# X_test = X_test/255 - 0.5
# X_val = X_val/255 - 0.5

print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
print(y_train.shape)
print(y_test.shape)
print(y_val.shape)
```

```python
def forward_prop(X, w1, b1, w2, b2):
    z1 = tf.add(X @ w1, b1)
    a1 = tf.nn.sigmoid(z1)
    z2 = tf.add(a1 @ w2, b2)
#     a2 = tf.nn.tanh(z2)
#     z3 = tf.add(a2 @ w3, b3)
#     a3 = tf.nn.sigmoid(z3)
#     z4 = tf.add(a3 @ w4, b4)
#     a4 = tf.nn.tanh(z4)
    
    return z2


def loss(z4, Y, w1, w2, lamda=0):
    j = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=z4, labels=Y)) \
    + 0.5 * lamda * (tf.nn.l2_loss(w1) + tf.nn.l2_loss(w2)) # + tf.nn.l2_loss(w3) + tf.nn.l2_loss(w4))
    return j
```

```python
import time
start = time.time()
alpha = 0.01
lamda = 0.001
layer1_size = 80
layer2_size = 10
layer3_size = 10
epochs = 500

tf.set_random_seed(777)
x = tf.placeholder(tf.float32, [None, n])
y = tf.placeholder(tf.float32, [None, n_classes])

w1 = tf.Variable(initial_value=tf.random_normal([n, layer1_size]), dtype=tf.float32)
# w2 = tf.Variable(initial_value=tf.random_normal([layer1_size, layer2_size]), dtype=tf.float32)
# w3 = tf.Variable(initial_value=tf.random_normal([layer2_size, layer3_size]), dtype=tf.float32)
w2 = tf.Variable(initial_value=tf.random_normal([layer1_size, n_classes]), dtype=tf.float32)
b1 = tf.Variable(initial_value=tf.random_normal([1, layer1_size]), dtype=tf.float32)
# b2 = tf.Variable(initial_value=tf.random_normal([1, layer2_size]), dtype=tf.float32)
# b3 = tf.Variable(initial_value=tf.random_normal([1, layer3_size]), dtype=tf.float32)
b2 = tf.Variable(initial_value=tf.random_normal([1, n_classes]), dtype=tf.float32)

z4 = forward_prop(x, w1, b1, w2, b2)

cost = loss(z4, y, w1, w2, lamda=lamda)
accuracy = 100 * tf.reduce_mean(tf.cast(tf.equal(tf.argmax(z4, axis=1), tf.argmax(y, axis=1)), tf.float32))

optimizer = tf.train.AdamOptimizer(learning_rate = alpha, beta1=0.9, beta2=0.999).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

j_history = []
j_val_history = []

for epoch in range(epochs):
    _, j, acc = sess.run([optimizer, cost, accuracy], feed_dict={x: X_train, y: y_train})
    j_val, acc_val = sess.run([cost, accuracy], feed_dict={x: X_val, y: y_val})
    
    j_history.append(j)
    j_val_history.append(j_val)
    print("Epoch: %d\tCost: %0.3f\tTrain Accuracy: %0.2f\tValidation Accuracy: %0.2f" % (epoch + 1, j, acc, acc_val))

train, = plt.plot(range(1, epochs + 1), j_history)
val, = plt.plot(range(1, epochs + 1), j_val_history)
plt.xlabel("No. of Iterations")
plt.ylabel("Cost")
plt.legend([train, val], ['Train Cost', 'Validation Cost'])
plt.show()

# parameters = sess.run([w1, b1, w2, b2, w3, b3, w4, b4])
# w1, b1, w2, b2, w3, b3, w4, b4 = parameters[0], parameters[1], parameters[2], parameters[3], parameters[4], parameters[5], parameters[6], parameters[7]

print("Training accuracy: %0.2f\tTest set accuracy: %0.2f\tValidation set accuracy: %0.2f"
      % (sess.run(accuracy, feed_dict={x: X_train, y: y_train}), 
        sess.run(accuracy, feed_dict={x: X_test, y: y_test}), 
        sess.run(accuracy, feed_dict={x: X_val, y: y_val})))
end = time.time()
print("Execution time: %0.2f s" % (end - start))
```

```python
acc_gross = (sess.run(accuracy, feed_dict={x: X_train, y: y_train}) 
       + sess.run(accuracy, feed_dict={x: X_test, y: y_test}) 
       + sess.run(accuracy, feed_dict={x: X_val, y: y_val})) / 3
print("Gross Accuracy: %0.2f %%" % (acc_gross))
print("Test Accuracy: %0.2f %%" % (sess.run(accuracy, feed_dict={x: X_test, y: y_test})))
```
